```python
from bs4 import BeautifulSoup as BS
import requests
import json
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population'
```

```python
data = requests.get(url)
print data
```

```python
soup = BS(data.text)
table = soup.find_all('table')
print len(table)
```

```python
#print table[0]
tr = table[0].find_all('tr')
```

```python
country = []

for ix in range(1, len(tr)):
    c_data = {}
    
    td = tr[ix].find_all('td')
    
    country_names = td[1].find_all('a')[0]
    country_name = country_names.text
    c_data['country'] = country_name
    
    population = td[-4].text
    c_data['population'] = population
    
    contrib = td[-2].text
    c_data['contribution'] = contrib
    
    country.append(c_data)
```

```python
print country
```

```python
data = {'data' : country}
print data
```

```python
y = json.dumps(data)
print len(y)

f = open('countries_and_population.json', 'w')
f.write(y)
f.close()
```
